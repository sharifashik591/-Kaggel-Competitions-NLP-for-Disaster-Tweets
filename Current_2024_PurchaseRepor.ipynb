### Sheet data Processing

```python
all_coll = {}
all_sheet_data = None
for index, sheetName in enumerate(filtered_sheet_names):

  # if index == round(len(filtered_sheet_names) / 4):
  if index in [5,10,15,20,25,30,40]:
    time.sleep(20)

  worksheet = gc.open_by_key(spreadsheet_key).worksheet(sheetName)
  df = getSheetDate(worksheet)
  print(sheetName)
  col = df.columns
  all_coll[sheetName] = col
  if all_sheet_data is None:
    all_sheet_data = df
  else:
    all_sheet_data = pd.concat([all_sheet_data, df], axis=0)

  # break

all_sheet_data
```

```python
df = all_sheet_data.copy()
df = df[(df['sl'] != '') & (df['sl'].notna()) & (df['sl'] != ' ') &
     (df['product name'] != '') & (df['product name'].notna()) &
     (df['total price'] != '') & (df['total price'].notna()) &
     (df['unit'] != '') & (df['unit'].notna()) &
     (df['variant'] != '') & (df['variant'].notna())]

df = df[['product id', 'product name', 'variant', 'unit', 'qty', 'unit price', 'total price', 'Date Name']]
```

```python
df[df['qty'] == '10$']
```

```python
df['qty'] = pd.to_numeric(df['qty'].astype(str).str.replace(r'[^\d.]', '', regex=True))

df['unit price'] = pd.to_numeric(df['unit price'].astype(str).str.replace(',', ''))

df['total price'] = df['qty'] * df['unit price']
```

## Convert Date

```python
from datetime import datetime

def convert_date_format(date_str):
  try:
    date_obj = datetime.strptime(date_str, '%d-%b-%y')
    return date_obj.strftime('%Y-%m-%d')
  except ValueError:
    return None  # Handle invalid date formats if needed

# Assuming 'Date Name' is the column containing the dates
df['Date Name'] = df['Date Name'].apply(convert_date_format)
```
